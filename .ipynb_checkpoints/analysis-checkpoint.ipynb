```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_theme()
sns.set(rc={'figure.figsize':(12,8)})
```

```python
df = pd.read_excel('out/output.xlsx', sheet_name = 0, index_col = 0, parse_dates = ['date', 'run_time']).fillna('')
df.head()
```

```python
agg_list = ['search_phrase', 'run_time', 'sections']

sums = df.groupby(
    by=agg_list
)[['money', 'count']].sum().rename(
    columns={'money':'sum_money','count':'sum_count'}
)

has_image = df.groupby(
    by=agg_list
)[['img_name']].apply(lambda x: (x != '').sum()).rename(
    columns={'img_name':'has_image'}
)

counts = df.groupby(
    by=agg_list
)[['date']].count().rename(
    columns={'date':'row_count'}
)

df_agg = counts.join(has_image).join(sums).reset_index()

df_agg['has_image_percent'] = df_agg['has_image'] / df_agg['row_count']

df_agg.sort_values(by='row_count', ascending=True).reset_index(drop=True)
```

```python
sns.barplot(data=df_agg, x="search_phrase", y="row_count")
```

```python
sns.barplot(data=df_agg, x="search_phrase", y="sum_money")
```
